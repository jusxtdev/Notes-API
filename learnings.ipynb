### Using status code and `JSONResponse`

```python
from fastapi import status
from fastapi.responses import JSONResponse

def get_note():
    status_code = status.HTTP_200_OK
    content = {'message' : 'new note'}
    return JSONResponse(content=content, status_code=status_code)
```

- If you have a pydantic model inside the 'content'
- Then `JSONResponse` will throw - `Object not JSON serializable` error     
To fix that - Use `jsonable_encoder()` 

```python

from fastapi import status
from fastapi.responses import JSONResponse
from fastapi.encoders import jsonable_encoder

def get_note():
    status_code = status.HTTP_200_OK
    content = {'message' : 'new note'}

    # ---
    content = jsonable_encoder(content)
    # ---
    return JSONResponse(content=content, status_code=status_code)
```
