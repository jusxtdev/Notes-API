- When working with api and want to return the time (like created at etc)
- you should return in __ISO-8601__ format
- `2025-01-20T14:35:22Z`

```python
from datetime import datetime

timestamp = datetime.now()      # ISO-8601 format
print(timestamp)
```

### Using status code and `JSONResponse`

```python
from fastapi import status
from fastapi.responses import JSONResponse

def get_note():
    status_code = status.HTTP_200_OK
    content = {'message' : 'new note'}
    return JSONResponse(content=content, status_code=status_code)
```
